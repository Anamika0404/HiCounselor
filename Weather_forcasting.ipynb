```python
#file path
df= pd.read_excel("weather_forecasting_dataset.xls")
df
```

# Subtask 1: Correct years for given data set

```python
# Replace all occurrences of single quotes with an empty string
df = df.applymap(lambda x: x.replace("'",""))

# Convert the date column to a datetime format
df['Date'] = pd.to_datetime(df['Date'],format= '%Y-%m-%d',errors = 'coerce' )
df['Date1'] = pd.to_datetime(df['Date1'],format= '%Y-%m-%d',errors = 'coerce' )

df
```

```python
#drop any rows with invalid dates
df = df[df['Date'].dt.date.astype('str').str.match(r'\d{4}-\d{2}-\d{2}')]
#reset the index of dataframe
df = df.reset_index(drop =True)
#extract the year from the "Date" coloumn
df['year']= df['Date'].dt.year
#initialize a flag to track wheather the year needs to be increased
increase_year =False
```

```python
#loop through the dataframe and update year needs to be increased

for i in range(1,len(df)):
    if df.loc[i,'Date'].month<df.loc[i-1,'Date'].month:
        increase_year = True
    if  increase_year:
        df.loc[i:,'year'] =df.loc[i:,'year'] + 1
        increase_year = False
```

```python
#create a new "Date" coloumn with the corrected year value
df['Date'] =pd.to_datetime(df['year'].astype(str) + df['Date'].dt.strftime('-%m-%d'))

#drop the year column as it is no longer needed
df =df.drop(columns = 'year')
```
